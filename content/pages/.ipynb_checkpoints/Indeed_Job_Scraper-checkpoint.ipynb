```python
from bs4 import BeautifulSoup as Soup
from urllib.request import urlopen
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


def results(query = 'data+scientist', location = 'Seattle', sort_by = 'date', binary_level = 1):
    pd.set_option('max_colwidth',500)    # to remove column limit (Otherwise, we'll lose some info)
    jobs = pd.DataFrame()

    # Get results
    soup_count = get_soup('http://www.indeed.com/jobs?q={}&jt=fulltime&sort={}&l={}'.format(query, sort_by, location))
    
    results_number = soup_count.find('div', attrs = {'id': 'searchCount'}).text
    number_of_results = int(results_number.split(sep = ' ')[-1].replace(',', ''))

    # Loop through the pages viewing 100 results at a time.
    i = int(number_of_results/100)
    for page_number in range(i + 1):
         
        soup_results = get_soup('http://www.indeed.com/jobs?q={}&jt=fulltime&sort={}&l={}&limit=100&start={}'.format(query, sort_by, location, str(100 * page_number)))
        results = soup_results.find_all('div', attrs={'data-tn-component': 'organicJob'})

        # trying to get each specific job information (such as company name, job title, urls, ...)
        for elem in results:
            comp_name = elem.find('span', attrs={'itemprop':'name'}).text.strip().capitalize()
            job_title = elem.find('a', class_='turnstileLink').attrs['title']
            #job_title = elem.find('a', attrs={'data-tn-element': "jobTitle"}).text.strip().capitalize()
            home_url = 'http://www.indeed.com'
            job_link = '%s%s' % (home_url,elem.find('h2', attrs={'class': 'jobtitle'}).find('a')['href'])
            #job_link = "https://www.indeed.com" + elem.find('h2', attrs={"class": "jobtitle"}).find('a')['href']
            job_addr = elem.find('span', attrs={'itemprop':'addressLocality'}).text
            job_posted = elem.find('span', attrs={'class': 'date'}).text
            job_id = elem.find('h2', attrs={'class': 'jobtitle'})['id']
            days_posted = return_days_posted(job_posted)
            
            
            """### Follow link to job posting and strip job description
            #soup_for_desc = Soup(urlopen(job_link).read(),'html.parser')
            #soup_p = soup_for_desc.findAll('p')
            #soup_ptrim = [str(s).replace('<p>', '').replace('</p>', '').replace('<span>', '').replace('</span>', '') for s in soup_p if 'id=' not in str(s) and 'class=' not in str(s) and 'href=' not in str(s) and '<b>' not in str(s) and 'consideration for employment without regard to' not in str(s) and 'lease try again' not in str(s)]
            #job_desc = '\n'.join([s for s in soup_ptrim if len(s) > 36])"""
                        
            
            comp_link_overall = elem.find('span', attrs={'itemprop':'name'}).find('a')
            if comp_link_overall != None: # if company link exists, access it. Otherwise, skip.
                comp_link_overall = '%s%s' % (home_url, comp_link_overall.attrs['href'])
            else: comp_link_overall = None

            # add a job info to our data frame
            jobs = jobs.append({'comp_name': comp_name, 'job_title': job_title, 
                            'job_link': job_link, 'job_posted': job_posted,
                            'overall_link': comp_link_overall, 'job_location': job_addr,
                            'overall_rating': None, 'wl_bal_rating': None, 
                            'benefit_rating': None, 'jsecurity_rating': None, 
                            'mgmt_rating': None, 'culture_rating': None,
                            'job_id': job_id, 'days_posted': days_posted
                            #'job_description': job_desc
                           }, ignore_index=True)

    def get_soup(url):
        return Soup(urlopen(url).read(), 'html.parser')
    
    def return_days_posted(job_posted):
            if job_posted == 'Just posted':
                return 'JN'
            elif job_posted == 'Today':
                return 'T'
            else:
        return job_posted.split()[0]
    
    # Remove re-posted jobs
    jobs.drop_duplicates(inplace = True)
    print('Finished')
    return jobs
```

```python
df = results()
```

```python
# InlineBackend.figure_formats : Set

#     Default: set([‘png’])

#     A set of figure formats to enable: ‘png’, ‘retina’, ‘jpeg’, ‘svg’, ‘pdf’.

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

sns.set(style="white")
#bins = np.append(['JN','T'],np.append(np.arange(1,31).astype(str),'30+'))
bins = np.append(['JN','T'],np.arange(1,31).astype(str))


ax = sns.countplot(x='days_posted',
                   data=df,
                   order=bins
                  )
ax.set(title='Frequency of Data Scientist Jobs Posts - Past 30 Days from May 1nd 2017', xlabel='Number of days since job posting', ylabel='Jobs Posted')
plt.show()
```
